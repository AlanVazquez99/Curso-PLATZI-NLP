```python
import nltk
from nltk.util import ngrams

nltk.download('book')
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.book import *

stop_words = stopwords.words('english')

import matplotlib.pyplot as plt
import numpy as np
import re 
```

```python
pattern = r'\w+(?:-\w+)*'
search = re.compile(pattern)

text = [word.lower() for word in text1 if search.search(word) if word not in stop_words]

print(text[:10])
len(text)
```

```python
bigrams = list(ngrams(text, 2))
trigrams = list(ngrams(text, 3))

print(len(bigrams))
print(bigrams[:10])

print(len(trigrams))
print(trigrams[:10])
```

```python
bigrams_frequency = FreqDist(bigrams)
trigrams_frequency = FreqDist(trigrams)

print(bigrams_frequency.most_common(10))
print(trigrams_frequency.most_common(10))
```

```python
plt.figure(figsize=(12, 5))
bigrams_frequency.plot(20)

plt.figure(figsize=(12, 5))
trigrams_frequency.plot(20)
```
